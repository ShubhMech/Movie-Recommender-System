```python
meta_more_than_min=meta[meta['vote_count']>meta['vote_count'].quantile(0.9)]
```

```python
meta_more_than_min['vote_average']
```

```python
def weighted_score(x,min__=min_,mean_vote=meta_more_than_min['vote_average'].mean()):
    voters= x['vote_count']
    avg_vote=x['vote_average']
    return (voters*avg_vote/(voters+min__)+(min__*mean_vote/(voters+min__)))
```

```python
meta_more_than_min['Weighted Score']= meta_more_than_min.apply(weighted_score, axis=1)
```

```python
meta_more_than_min
```

```python
meta__=meta_more_than_min.sort_values(by='Weighted Score')
```
